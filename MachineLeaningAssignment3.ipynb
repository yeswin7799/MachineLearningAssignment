```python
missing_counts = df.isnull().sum()

print("Missing values per column:")
print(missing_counts)

if missing_counts.sum() == 0:
    print("\nNo imputation needed the dataset contains no missing values.")
else:
    print("\nMissing values detected. Imputing with column mean...")
    missing_indices = df[df.isnull().any(axis=1)].index.tolist()
    print("Indices with missing values:", missing_indices)
    
    df.fillna(df.mean(), inplace=True)
    print("Imputation completed using column means.")
```

```python
df
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
df['MEDV_bin'] = pd.cut(
    df['MEDV'],
    bins=[0, 10, 20, 30, 40, 50],
    labels=[1, 2, 3, 4, 5]
)
```

```python
print("Distribution of MEDV bins:")
print(df['MEDV_bin'].value_counts())
```

```python

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=240)

for train_idx, test_idx in split.split(df, df['MEDV_bin']):
    train_set = df.loc[train_idx]
    test_set = df.loc[test_idx]

train_set = train_set.drop('MEDV_bin', axis=1)
test_set = test_set.drop('MEDV_bin', axis=1)


```

```python
X_train_raw = train_set.drop("MEDV", axis=1)
y_train = train_set["MEDV"]

X_test_raw = test_set.drop("MEDV", axis=1)
y_test = test_set["MEDV"]


print("\nTrain set size:", X_train_raw.shape)
print("Test set size:", X_test_raw.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train_raw)

# Transform training and test sets
X_train_std = scaler.transform(X_train_raw)
X_test_std = scaler.transform(X_test_raw)

X_train_std = pd.DataFrame(X_train_std, columns=X_train_raw.columns, index=X_train_raw.index)
X_test_std = pd.DataFrame(X_test_std, columns=X_test_raw.columns, index=X_test_raw.index)

print("Standardization completed.")
print("\nTraining set (standardized) preview:")
display(X_train_std.head())
```

```python

print("\nTesting set (standardized) preview:")
display(X_test_std.head())
```

```python
from sklearn.linear_model import LinearRegression

X_train_RM = X_train_raw[['RM']]   
X_test_RM = X_test_raw[['RM']]

# Create and train the model
lin_reg = LinearRegression()
lin_reg.fit(X_train_RM, y_train)
```

```python
# Get parameters
beta_0 = lin_reg.intercept_
beta_1 = lin_reg.coef_[0]

print("Linear Regression on RM")
print(f"Intercept (Î²0): {beta_0:.4f}")
print(f"Slope (Î²1): {beta_1:.4f}")

# Model form
print(f"\nModel:  MEDV = {beta_0:.4f} + {beta_1:.4f} * RM")
```

```python
plt.figure(figsize=(7,5))

plt.scatter(X_train_RM, y_train, alpha=0.6, label='Training data')

# Regression line values
RM_range = np.linspace(X_train_RM.min(), X_train_RM.max(), 100)
RM_range = RM_range.reshape(-1, 1)
MEDV_pred_line = lin_reg.predict(RM_range)

# Plot regression line
plt.plot(RM_range, MEDV_pred_line, color='red', linewidth=2, label='Regression Line')

# Labels and title
plt.xlabel("RM (Average number of rooms per dwelling)")
plt.ylabel("MEDV (Median home value in $1000s)")
plt.title("Linear Regression Fit: RM vs MEDV")
plt.legend()

# Annotate intercept
plt.text(
    X_train_RM.min(), 
    beta_0, 
    f"Intercept = {beta_0:.2f}", 
    color='red',
    fontsize=10
)

plt.grid(True)
plt.show()
```

```python
y_pred_test = lin_reg.predict(X_test_RM)

plt.figure(figsize=(7,5))
plt.scatter(y_test, y_pred_test, alpha=0.6, color='teal', label='Predictions')

# Perfect prediction reference line
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         color='red', linestyle='--', linewidth=2, label='Perfect Fit (y = x)')

plt.xlabel("True MEDV")Z
plt.ylabel("Predicted MEDV")
plt.title("Predicted vs True MEDV (Linear Regression on RM)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np


MAE = mean_absolute_error(y_test, y_pred_test)
MSE = mean_squared_error(y_test, y_pred_test)
RMSE = np.sqrt(MSE)
R2 = r2_score(y_test, y_pred_test)

print("Evaluation Metrics (Test Set):")
print(f"MAE:  {MAE:.4f}")
print(f"MSE:  {MSE:.4f}")
print(f"RMSE: {RMSE:.4f}")
print(f"RÂ²:   {R2:.4f}")
```

```python
from mpl_toolkits.mplot3d import Axes3D

beta_0_range = np.linspace(beta_0 - 20, beta_0 + 20, 50)
beta_1_range = np.linspace(beta_1 - 5, beta_1 + 5, 50)

B0, B1 = np.meshgrid(beta_0_range, beta_1_range)

MSE_surface = np.zeros_like(B0)

X = X_train_RM.values.flatten()
y = y_train.values

for i in range(B0.shape[0]):
    for j in range(B0.shape[1]):
        y_pred = B0[i, j] + B1[i, j] * X
        MSE_surface[i, j] = np.mean((y - y_pred)**2)

# Plot 3D surface
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(B1, B0, MSE_surface, cmap='viridis', edgecolor='none')
ax.set_xlabel('Slope (Î²1)')
ax.set_ylabel('Intercept (Î²0)')
ax.set_zlabel('MSE')
ax.set_title('3D Error Surface for Linear Regression on RM')

plt.show()

```

```python

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Define the model
model = keras.Sequential([
    layers.Input(shape=(13,)),           
    layers.Dense(13, activation='relu'), 
    layers.Dense(8, activation='relu'),  
    layers.Dense(1, activation='linear') 
])

# Display model summary
model.summary()
```

```python

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),  
    loss='mse',                                            
    metrics=['mse', 'mae']                                  
)

history = model.fit(
    X_train_std, y_train,
    epochs=25,
    batch_size=10,
    validation_split=0.1,
    verbose=1
)

print("Training complete.")
```

```python

from sklearn.metrics import r2_score

y_pred_nn = model.predict(X_test_std).flatten()

test_mse = mean_squared_error(y_test, y_pred_nn)
test_mae = mean_absolute_error(y_test, y_pred_nn)
test_r2  = r2_score(y_test, y_pred_nn)

print("Neural Network Test Performance:")
print(f"MSE:  {test_mse:.4f}")
print(f"MAE:  {test_mae:.4f}")
print(f"RÂ²:   {test_r2:.4f}")
```

```python
plt.figure(figsize=(7,5))

# Plot training and validation MSE
plt.plot(history.history['loss'], label='Training MSE', linewidth=2)
plt.plot(history.history['val_loss'], label='Validation MSE', linewidth=2)

plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.title("Learning Curves: Training vs Validation MSE")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.scatter(y_test, y_pred_nn, alpha=0.6, color='purple', label='Predictions')

# Perfect fit reference line
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         color='red', linestyle='--', linewidth=2, label='Perfect Fit (y = x)')

plt.xlabel("True MEDV")
plt.ylabel("Predicted MEDV")
plt.title("Neural Network: Predicted vs Actual MEDV")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.svm import SVR

svr_model = SVR(
    kernel='linear',
    C=1.0,
    epsilon=0.1
)

svr_model.fit(X_train_std, y_train)

print("SVR model trained successfully.")
```

```python
beta_0_svr = svr_model.intercept_[0]
beta_svr = svr_model.coef_.ravel()   # shape: (13,)

feature_names = X_train_std.columns

print("SVR (linear kernel) coefficients:")
for name, coef in zip(feature_names, beta_svr):
    print(f"{name}: {coef:.3f}")

print(f"\nIntercept (Î²0): {beta_0_svr:.3f}")

terms = [f"{beta_svr[i]:.3f} * {feature_names[i]}" for i in range(len(feature_names))]
equation = " + ".join(terms)

print("\nRegression Equation (standardized features):")
print(f"Y = {beta_0_svr:.3f} + " + equation + " + Îµ")
```

```python

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Predict on test set
y_pred_svr = svr_model.predict(X_test_std)

# Compute metrics
svr_mae = mean_absolute_error(y_test, y_pred_svr)
svr_mse = mean_squared_error(y_test, y_pred_svr)
svr_rmse = np.sqrt(svr_mse)
svr_r2 = r2_score(y_test, y_pred_svr)

print("SVR Test Metrics:")
print(f"MAE:  {svr_mae:.4f}")
print(f"MSE:  {svr_mse:.4f}")
print(f"RMSE: {svr_rmse:.4f}")
print(f"RÂ²:   {svr_r2:.4f}")
```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))

# Scatter plot of true vs predicted values
plt.scatter(y_test, y_pred_svr, alpha=0.6, color='darkorange', label='SVR Predictions')

# Perfect prediction reference line (y = x)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         color='blue', linestyle='--', linewidth=2, label='Perfect Fit (y = x)')

plt.xlabel("True MEDV")
plt.ylabel("Predicted MEDV")
plt.title("SVR: Predicted vs Actual MEDV")
plt.legend()
plt.grid(True)
plt.show()
```
